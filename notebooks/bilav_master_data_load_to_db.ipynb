```python
# seperate the datasets in separate files
def split_files(dir_path, file_name, asof_date, dict_columns):
    file_name = os.path.join(dir_path, file_name)
    print(file_name)

    with open (file_name, encoding='utf-8-sig') as f:
        r = f.readlines()

    dict_filenames = {}
    for key, val in dict_columns.items():
        print(f'Splitting data for {key}')
        is_column_set = 0
        for i in range(len(r)):
            # setting column header
            if is_column_set:
                row = r[i]
                data_type = r[i].split('|')[0]
            else:
                row = '|'.join(val)
                row += '\n'
                is_column_set = 1
                data_type = key

            if key == data_type.upper():
                # write the row in the file
                filename = "bilav_" + data_type.upper() + "_" + asof_date.strftime("%d%m%Y") + ".csv"
                filename = os.path.join(dir_path, filename)
                with open(filename,'a') as f:
                    f.write(row)

                # capture new filenames
                if not (key in dict_filenames):
                    dict_filenames[key] = filename

    return dict_filenames
```

```python
def get_fi_data_df(file_name, asof_date):
    print(file_name)
    if os.path.isfile(file_name):
        print('Reading the file into df...')
        df = pd.read_csv(file_name, sep ='|', header=0, encoding='unicode_escape') # encoding='utf-8-sig'   encoding= 'unicode_escape'
    else:
        df = pd.DataFrame()
    return df
```
